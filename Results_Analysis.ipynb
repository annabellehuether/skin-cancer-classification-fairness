# Results Analysis and Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
```

```python
metric_df = pd.read_csv('Evaluation_Metrics.csv')
```

```python
metric_df
```

```python
experiments = ['1-1' ,'1-2', '1-3', '2-1', '2-2' ,'2-3', '2-4' ,'3-1', '3-2', '3-3' ,'3-4', '4-1', '4-2','4-3' ,'4-4']
metrics1 = metric_df['InceptionV3 HAM Macro Avg F1']
metrics2 = metric_df['MobileNetV2 HAM Macro Avg F1']

bar_width = 0.35

x = np.arange(len(experiments))

plt.bar(x - bar_width/2, metrics1, bar_width, label='InceptionV3', color='dodgerblue')
plt.bar(x + bar_width/2, metrics2, bar_width, label='MobileNetV2', color='orange')

averages = [(m1 + m2) / 2 for m1, m2 in zip(metrics1, metrics2)]

for i, avg in enumerate(averages):
    plt.text(i, max(metrics1[i], metrics2[i]), f'{avg:.2f}', ha='center', va='bottom', fontsize=8)

plt.title('Macro Average F1 on HAM10000 Test Set', fontweight="bold")
plt.xlabel('Experiment')
plt.ylabel('Macro Average F1')
plt.ylim(0.3, 0.8)
plt.xticks(x, experiments, rotation=45)

plt.legend()
plt.tight_layout()
plt.show()
```
